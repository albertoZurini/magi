# **The cart and pole system**

Consider the cart-and-pole system here depicted:
<div>
<img src="https://drive.google.com/uc?export=view&id=1Yu5zZ57OEEYZ678yWLaU7h-Wq3M9vV9E" width="200">
</div>
whose dynamics follows the following equations:

\begin{align}
  \ddot{\theta}&=\frac{g \sin(\theta)\,+\,\cos(\theta) \left[ \frac{-F\,-\,m_p\,l\,\dot{\theta}^2 \sin \left(\theta\right)}{m_c\,+\,m_p}\right]-\frac{\mu_p\dot{\theta}}{m_p\,l}}{l\,\left[\frac{4}{3}-\frac{m_p\,cos^2\left(\theta\right)}{m_c\,+\,m_p}\right]}
\end{align}
>>>>>>>>>>>>>>>>>>>>>>>>$(*)$

\begin{align}
  \ddot{p}_c&=\frac{F\,+\,m_p\,l\,\left[\dot{\theta}^2\,\sin\left(\theta\right)\,-\,\ddot{\theta}\,\cos(\theta)\right]}{m_c\,+\,m_p}
\end{align}

where $g = 9.8\,\text{m/sec}$ is the gravitational acceleration, $m_c = 1\,\text{kg}$ is the cart mass, $m_p = 0.1\,\text{kg}$ is the pole mass, $l = 0.5\,\text{m}$ is the half-pole length, $\mu_p = 0.000002$ is the pole on cart friction coefficient, and $F:|F|\leq 10\,\text{N}$ is the force applied to cart's center of mass.

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
from scipy.io import loadmat
import jax.numpy as jnp
from jax import jacfwd
from jax import config
from scipy.optimize import fsolve
from IPython.display import HTML
import matplotlib.animation as animation
from matplotlib import rc
import matplotlib

matplotlib.rcParams['animation.embed_limit'] = 2**128
rc('animation', html='jshtml')
config.update("jax_enable_x64", True)
```

```python
class InvertedPendulum:
  def __init__(self, m_c, m_p, g, l, u_p):
    self.m_c = m_c
    self.m_p = m_p
    self.g = g
    self.l = l
    self.u_p = u_p

  ############################################################################
  ## Computation of the non-linear system
  ############################################################################

  def compute(self, t, x: np.array, u: float) -> jnp.array:
    dx = [None for _ in range(4)]
    dx[0] = x[1]
    dx[1] = (self.g*jnp.sin(x[0]) + jnp.cos(x[0])*((-u -self.m_p*self.l*jnp.sin(x[0])*x[1]**2)/(self.m_c + self.m_p)) - (self.u_p*x[1])/(self.m_p*self.l))/(self.l*(4/3 - (self.m_p*jnp.cos(x[0])**2)/(self.m_c + self.m_p)))
    dx[2] = x[3]
    dx[3] = (u + self.m_p*self.l*(jnp.sin(x[0])*x[1]**2 - dx[1]*jnp.cos(x[0])))/(self.m_c + self.m_p)

    return jnp.asarray(dx)

  ############################################################################
  ## Computation of the linearized system
  ############################################################################

  def lin_compute(self, t, x: np.array, u: float, A: jnp.array, B: jnp.array) -> jnp.array:
    dx = jnp.dot(A, x) + jnp.dot(B, u)
    return dx

  ############################################################################
  ## Linearization of the system
  ############################################################################

  def linearize(self, x_eq: np.array, u_eq: float) -> tuple:
    A = jacfwd(lambda x: self.compute(0, x, u_eq))(x_eq)
    B = jacfwd(lambda u: self.compute(0, x_eq, u))(u_eq)

    B = B.reshape(B.shape[0], 1)

    return (A, B)

  ############################################################################
  ## System rendering
  ############################################################################

  def render_cart_pole(self, time: float, state_evolution: np.array, sample_time: float):
    fig, ax = plt.subplots()

    cart_position = state_evolution[2, :]
    pole_angle = state_evolution[0, :]

    cart_width = 0.3
    cart_height = 0.2

    cart = plt.Rectangle((cart_position[0] - cart_width/2, 0), cart_width, cart_height, color='blue')
    pole_length = self.l*2
    pole = plt.Line2D([cart_position[0], cart_position[0] + pole_length*np.sin(pole_angle[0])],
                     [cart_height, cart_height + pole_length*np.cos(pole_angle[0])], color='black', linewidth=2)
    ax.set(xlim=[-2, 2], ylim=[-2, 2], xlabel='Position X [m]', ylabel='Position Y [m]')

    ax.grid()

    ax.add_patch(cart)
    ax.add_line(pole)

    def update(frame):
      cart.set_x(cart_position[frame] - cart_width/2)
      pole.set_data([cart_position[frame], cart_position[frame] + pole_length*np.sin(pole_angle[frame])],
                     [cart_height, cart_height + pole_length*np.cos(pole_angle[frame])])
      return (cart, pole)


    ani = animation.FuncAnimation(fig=fig, func=update, frames=state_evolution.shape[1], interval=sample_time*1000)
    plt.close()

    return ani
```

```python
def simulate(sys: InvertedPendulum, x_init: np.array, u_seq: np.array, sample_time: float) -> tuple: # simulation of the non linear system
  t_eval = np.linspace(0, sample_time*u_seq.shape[0], u_seq.shape[0] + 1) # creation of the time interval vector
  result = np.empty((x_init.shape[0], t_eval.shape[0])) # creation of the nd-array to store the results
  result[:, 0] = x_init

  # simulation
  for i in range(0, t_eval.shape[0] - 1):
    tspan = [t_eval[i], t_eval[i + 1]] # interval in which we are intereseted
    sol = solve_ivp(fun=lambda t, x: sys.compute(t, x, u_seq[i]), t_span=tspan, y0=x_init, t_eval=tspan, method='RK45', rtol=1e-3, atol=1e-6) # numerical integration
    x_init = sol.y[:, 1] # getting only the solution (not other stuff)
    result[:, i+1] = x_init # storing the solution
  return (t_eval, result)
```

```python
def lin_simulate(sys: InvertedPendulum, x_init: np.array, u_seq: np.array, sample_time: float, A: jnp.array, B: jnp.array) -> tuple: # simulation of the linear system
  t_eval = np.linspace(0, sample_time*u_seq.shape[0], u_seq.shape[0] + 1)
  result = np.empty((x_init.shape[0], t_eval.shape[0]))
  result[:, 0] = x_init

  # simulation
  for i in range(0, t_eval.shape[0] - 1):
    tspan = [t_eval[i], t_eval[i + 1]]
    sol = solve_ivp(fun=lambda t, x: sys.lin_compute(t, x, u_seq[i], A=A, B=B), t_span=tspan, y0=x_init, t_eval=tspan, method='RK45', rtol=1e-3, atol=1e-6) # the only difference with the function above is the usage of the sys.lin_compute function
    x_init = sol.y[:, 1]
    result[:, i+1] = x_init
  return (t_eval, result)
```

```python
def plot_solutions(t, solutions): # function able to plot the solutions
  # Create a figure with 2x2 subplots
  fig, axes = plt.subplots(1, 2, figsize=(10, 3))
  # Subplot 1 (left)
  axes[0].plot(t, solutions[0, :], label=r'theta')
  axes[0].set_title(r'$\theta(t)$')
  axes[0].legend()
  # Subplot 2 (right)
  axes[1].plot(t, solutions[0, :], label='theta')
  axes[1].plot(t, solutions[1, :], label='d(theta)')
  axes[1].plot(t, solutions[2, :], label='p')
  axes[1].plot(t, solutions[3, :], label='d(p)')
  axes[1].set_title(r'$\theta(t)$, $\dot{\theta}(t)$, $p(t)$, $\dot{p}(t)$')
  axes[1].legend()
  # Adjust layout
  plt.tight_layout()
  # Show the plot
  plt.show()
```

## **State and input vector definition**
Here we have the definition of the **state vector** and the **domain** in which it's defined:
\begin{align}
\begin{bmatrix}
x_1\\
x_2\\
x_3\\
x_4
\end{bmatrix} =
\begin{bmatrix}
\theta\\
\dot{\theta}\\
p\\
\dot{p}
\end{bmatrix}
\in \mathcal{A}
&&
u = F \in \left[-10, 10\right]
\end{align}
### **Note on the domain**
$\mathcal{A} âŠ‚ \mathbb{R}^4$ this is due to the fact that since $u$ is our only input we have an inplicit constraint to the state variable $x$.

- $x_1$ represents the pole angular position
- $x_2$ represents the pole angular velocity
- $x_3$ represents the cart position
- $x_4$ represents the cart velocity

### **Re-written equations**
\begin{align}
\begin{bmatrix}
\dot{x}_1\\
\dot{x}_2\\
\dot{x}_3\\
\dot{x}_4
\end{bmatrix} = \begin{bmatrix}
x_2\\
\frac{g \sin(x_1)\,+\,\cos(x_1) \left[ \frac{-F\,-\,m_p\,l\,x_2^2 \sin \left(x_1\right)}{m_c\,+\,m_p}\right]-\frac{\mu_p x_2}{m_p\,l}}{l\,\left[\frac{4}{3}-\frac{m_p\,cos^2\left(x_1\right)}{m_c\,+\,m_p}\right]} \\
x_4\\
\frac{F\,+\,m_p\,l\,\left[x_2^2\,\sin\left(x_1\right)\,-\,\dot{x}_2\,\cos(x_1)\right]}{m_c\,+\,m_p}
\end{bmatrix}
\end{align}

## **Equilibrium state**
As we can see from the simulation the equilibrium states are:

\begin{align*}
\begin{bmatrix}
\theta\\
\dot{\theta}\\
p\\
\dot{p}
\end{bmatrix}
=
\begin{bmatrix}
0, \pi \\
0\\
\mathbb{R}\\
0
\end{bmatrix}
\end{align*}

This is due to the fact that the equilibrium state does not depend on the position of the cart and pole, but from the fact that it's not moving. The equilibrium states are obtained by:

\begin{align*}
\begin{bmatrix}
\dot{\theta}\\
\ddot{\theta}\\
\dot{p}\\
\ddot{p}
\end{bmatrix}
=
\begin{bmatrix}
0\\
0\\
0\\
0
\end{bmatrix}
\end{align*}

Answer: (Arguing the answer)
