**The Tabular value-fucntion methods on the cart and pole system**

Consider the cart-and-pole system here depicted:
<div>
<img src="https://drive.google.com/uc?export=view&id=1Yu5zZ57OEEYZ678yWLaU7h-Wq3M9vV9E" width="200">
</div>
whose dynamics follows the following equations:

\begin{align}
  \ddot{\theta}&=\frac{g \sin(\theta)\,+\,\cos(\theta) \left[ \frac{-F\,-\,m_p\,l\,\dot{\theta}^2 \sin \left(\theta\right)}{m_c\,+\,m_p}\right]-\frac{\mu_p\dot{\theta}}{m_p\,l}}{l\,\left[\frac{4}{3}-\frac{m_p\,cos^2\left(\theta\right)}{m_c\,+\,m_p}\right]}
\end{align}
>>>>>>>>>>>>>>>>>>>>>>>>$(*)$

\begin{align}
  \ddot{p}_c&=\frac{F\,+\,m_p\,l\,\left[\dot{\theta}^2\,\sin\left(\theta\right)\,-\,\ddot{\theta}\,\cos(\theta)\right]}{m_c\,+\,m_p}
\end{align}

where $g = 9.8\,\text{m/sec}$ is the gravitational acceleration, $m_c = 1\,\text{kg}$ is the cart mass, $m_p = 0.1\,\text{kg}$ is the pole mass, $l = 0.5\,\text{m}$ is the half-pole length, $\mu_p = 0.000002$ is the pole on cart friction coefficient.

```python
class InvertedPendulum:
  def __init__(self, m_c, m_p, g, l, u_p):
    self.m_c = m_c
    self.m_p = m_p
    self.g = g
    self.l = l
    self.u_p = u_p

  ############################################################################
  ## Computation of the non-linear system
  ############################################################################

  def compute(self, t, x: np.array, u: float) -> tuple:
    dx = [None for _ in range(4)]
    dx[0] = x[1]
    dx[1] = (self.g*jnp.sin(x[0]) + jnp.cos(x[0])*((-u -self.m_p*self.l*jnp.sin(x[0])*x[1]**2)/(self.m_c + self.m_p)) - (self.u_p*x[1])/(self.m_p*self.l))/(self.l*(4/3 - (self.m_p*jnp.cos(x[0])**2)/(self.m_c + self.m_p)))
    dx[2] = x[3]
    dx[3] = (u + self.m_p*self.l*(jnp.sin(x[0])*x[1]**2 - dx[1]*jnp.cos(x[0])))/(self.m_c + self.m_p)

    dx = jnp.asarray(dx)

    return dx



  def render_cart_pole(self, time: float, state_evolution: np.array, sample_time: float):
    fig, ax = plt.subplots()

    cart_position = state_evolution[2, :]
    pole_angle = state_evolution[0, :]

    cart_width = 0.3
    cart_height = 0.2

    cart = plt.Rectangle((cart_position[0] - cart_width/2, 0), cart_width, cart_height, color='blue')
    pole_length = self.l*2
    pole = plt.Line2D([cart_position[0], cart_position[0] + pole_length*np.sin(pole_angle[0])],
                     [cart_height, cart_height + pole_length*np.cos(pole_angle[0])], color='black', linewidth=2)
    ax.set(xlim=[-2, 2], ylim=[-2, 2], xlabel='Position X [m]', ylabel='Position Y [m]')

    ax.grid()

    ax.add_patch(cart)
    ax.add_line(pole)

    def update(frame):
      cart.set_x(cart_position[frame] - cart_width/2)
      pole.set_data([cart_position[frame], cart_position[frame] + pole_length*np.sin(pole_angle[frame])],
                     [cart_height, cart_height + pole_length*np.cos(pole_angle[frame])])
      return (cart, pole)


    ani = animation.FuncAnimation(fig=fig, func=update, frames=state_evolution.shape[1], interval=sample_time*1000)
    plt.close()

    return ani
```

```python
def plot_solutions(t, solutions):
  # Create a figure with 2x2 subplots
  fig, axes = plt.subplots(1, 2, figsize=(10, 3))
  # Subplot 1 (left)
  axes[0].plot(t, solutions[0, :], label=r'theta')
  axes[0].set_title(r'$\theta(t)$')
  axes[0].legend()
  # Subplot 2 (right)
  axes[1].plot(t, solutions[0, :], label='theta')
  axes[1].plot(t, solutions[1, :], label='d(theta)')
  axes[1].plot(t, solutions[2, :], label='p')
  axes[1].plot(t, solutions[3, :], label='d(p)')
  axes[1].set_title(r'$\theta(t)$, $\dot{\theta}(t)$, $p(t)$, $\dot{p}(t)$')
  axes[1].legend()
  # Adjust layout
  plt.tight_layout()
  # Show the plot
  plt.show()
```

1. Given the $X$ set, select centers and standard deviation of RBFs and create a code that given a state $x$ returns $\phi\left(x,\cdot\right)$.

```python
dstate_space = [
    np.linspace(-np.pi/8, np.pi/8, 6),
    np.linspace(-1, 1, 4),
    np.linspace(-3, 3, 4),
    np.linspace(-2, 2, 4)
] # discretization of the state

dstate_space_list = [dstate_space[i].tolist() for i in range(len(dstate_space))] # discretized state space list

space_non_terminal_intervals = [
    (-np.pi/8, np.pi/8),
    (-1, 1),
    (-2.4, 2.4),
    (-2, 2)
] # non terminal state intervals
```

```python
rbf_centers = np.array(list(product(*dstate_space_list))) # rbf centers is the cartesian product of the dstate_space_list
```

```python
inputs = np.asarray([-10, 10])

weights = np.zeros((len(rbf_centers), len(inputs)))
weights.shape
```

```python
def rbf_func(state, means, std): # this is a function that returns the rbf calculated for every center
  state = state.reshape(-1, state.shape[0])
  state = np.repeat(state, len(means), axis=0)
  return np.exp(- (1/(2*std**2)) * (LA.norm((state - means), axis=1)**2))

means = rbf_centers # setting mean and std
std = 2/3

rbf_func(state=np.array([np.pi/13, 0, 0.2, 0]), means=means, std=std).shape
```

```python
def q_func(features, input): # computation of the q function
  input_idx = np.digitize(input, inputs) - 1
  return np.dot(weights[:, input_idx], rbf_func(features, means=means, std=std))

q_func(features=np.array([np.pi/13, 0, 0.2, 0]), input=-10)
```

```python
def random_state(): # generation of a random state in the non terminal bounds
  state = [np.random.uniform(low=space_non_terminal_intervals[i][0], high=space_non_terminal_intervals[i][1]) for i in range(len(space_non_terminal_intervals))]
  return np.asarray(state)

random_state()
```

```python
def eps_greedy_policy(epsilon, state): # here I define the function to apply the epsilon greedy policy
  greedy_vs_explore = np.random.uniform(0,1)

  if greedy_vs_explore > epsilon: # Exploiting (lower it is the more we are exploiting)
    u = max(inputs, key=lambda input: q_func(state, input)) # with this we return the input that maximizes the q_func
  else: # Exploring
    u = inputs[np.random.randint(inputs.shape[0])]

  u_idx = np.digitize(u, inputs) - 1 # here I retrive the index of the input, so [-10, 10] -> [0, 1]

  return u, u_idx

eps_greedy_policy(0.2, random_state())
```

```python
MAX_ITERATIONS = 1000

def get_rewarded_and_check_done(state, iteration): # here I check if the entered a terminal state and compute the rewards for the possible cases

    reward = 1
    done = False

    if iteration > MAX_ITERATIONS: # if we survived for 1000 steps we can terminate the episode
        reward = 0
        done = True
    elif abs(state[0]) > space_non_terminal_intervals[0][1]: # if the angle isn't in the range specified above we can terminate the episode
        reward = 0
        done = True
    elif abs(state[2]) > space_non_terminal_intervals[2][1]: # if the position isn't in the range specified above we can terminate the episode
        reward = 0
        done = True

    return reward, done

get_rewarded_and_check_done(random_state(), 10)
```

```python
ODE = jitcode(f)
ODE.set_integrator("dop853")


def step(state, input, iteration): # function that computes the next state, the reward and if we entered a terminal state
    in_vec = np.hstack((state, input)) # creation of the 5 element input

    ODE.set_initial_value(in_vec,0.0) # integration with the solver (since the step is alwais 0.02 we can integrate between [0, 0.02])
    sol = ODE.integrate(0.02)
    new_state = sol[:4]

    reward, done = get_rewarded_and_check_done(new_state, iteration) # computing the reward and if we entered a terminal state


    return new_state, reward, done

x0 = random_state()

x0, step(x0, 10, 10)
```

2. Create a code that applies Q-Learning algorithm with function approximation


*Provide the code*:



```python
alpha = 0.3 # setting the parameters for q and sarsa learning
gamma = 0.9

epsilon = 0.9

min_epsilon = 0.2

decay = 1e-4
```

```python
cum_rewards = []
rewards = []

for episode in tqdm(range(episodes)):
  x_init = random_state() # generation of a random state

  done = False
  iteration = 0

  cum_reward = 0

  while not done:
    u, u_idx = eps_greedy_policy(epsilon, x_init) # generation of epsilon greedy policy input

    new_state, reward, done = step(x_init, u, iteration) # computation of new state, reward and check if we entered a terminal state

    cum_reward += pow(gamma, iteration)*reward # compuation of the discounted comulative reward

    fq_estim = q_func(new_state, max(inputs, key=lambda input: q_func(new_state, input))) # esitimation of the future value of the Q function

    gradient = rbf_func(x_init, means=means, std=std) # the gradient in the linear case is just the rbf function
    weights[:, u_idx] = weights[:, u_idx] + np.dot(alpha*(reward + gamma*fq_estim - q_func(x_init, u)), gradient) # updating the weights
    x_init = new_state # resetting the state

    iteration += 1

    cum_rewards.append(cum_reward)
    rewards.append(cum_reward)

  epsilon = max(min_epsilon, (1 - min_epsilon)*np.exp(-decay*episode))

  if episode % 5000 == 0:
      print(f"{episode}: {np.average(cum_rewards)}, alpha={alpha}, epsilon: {epsilon}")
      cum_rewards = []
```

```python
plt.plot(np.convolve(rewards,np.ones(5000)/5000,mode='valid')) # plotting running average of the cumulated discounted reward (because from the rewards we don't understand anything)
```

```python
# Simulation with only greedy policy

x0 = np.array([-np.pi/18, 0, 0.1, 0])

t_eval = np.linspace(0, 20, 1000, endpoint=False)

result = np.empty((x0.shape[0], t_eval.shape[0]))
result[:, 0] = x0

u_seq = np.empty(t_eval.shape[0] - 1)

for i in range(t_eval.shape[0] - 1):
    u, _ = eps_greedy_policy(0, x0)
    u_seq[i] = u
    new_state, reward, _ = step(x0, u, i)

    result[:, i+1] = new_state

    x0 = new_state

plot_solutions(t_eval, result)
```

```python
weights = np.zeros((len(means), len(inputs))) # resetting weights
```

```python
cum_rewards = []
rewards = []

for episode in tqdm(range(episodes)): # here we have the sarsa algorithm, the difference from the the Q algorithm is just in the estimation of the future Q
  x_init = random_state()

  done = False
  iteration = 0

  cum_reward = 0

  u, u_idx = eps_greedy_policy(epsilon, x_init) # input generated with epsilon greedy policy

  while not done:

    new_state, reward, done = step(x_init, u, iteration) # computation of the new state, reward and check if the system is in a terminal state

    cum_reward += pow(gamma, iteration)*reward # computation of the discounted cumulative reward

    u_new, u_new_idx = eps_greedy_policy(epsilon, new_state) # computation of the new input based on the new state

    gradient = rbf_func(x_init, means=means, std=std) # computation of the gradient
    weights[:, u_idx] = weights[:, u_idx] + np.dot(alpha*(reward + gamma*q_func(new_state, u_new) - q_func(x_init, u)), gradient) # update of the weigths

    x_init = new_state #  resetting of the initial state and input
    u = u_new
    u_idx = u_new_idx

    iteration += 1

    cum_rewards.append(cum_reward)
    rewards.append(cum_reward)

  epsilon = max(min_epsilon, (1 - min_epsilon)*np.exp(-decay*episode))

  if episode % 5000 == 0:
      print(f"{episode}: {np.average(cum_rewards)}, alpha={alpha}, epsilon: {epsilon}")
      cum_rewards = []
```
