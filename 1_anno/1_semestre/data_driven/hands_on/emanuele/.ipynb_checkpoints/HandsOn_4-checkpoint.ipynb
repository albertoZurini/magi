**The Tabular value-fucntion methods on the cart and pole system**

Consider the cart-and-pole system here depicted:
<div>
<img src="https://drive.google.com/uc?export=view&id=1Yu5zZ57OEEYZ678yWLaU7h-Wq3M9vV9E" width="200">
</div>
whose dynamics follows the following equations:

\begin{align}
  \ddot{\theta}&=\frac{g \sin(\theta)\,+\,\cos(\theta) \left[ \frac{-F\,-\,m_p\,l\,\dot{\theta}^2 \sin \left(\theta\right)}{m_c\,+\,m_p}\right]-\frac{\mu_p\dot{\theta}}{m_p\,l}}{l\,\left[\frac{4}{3}-\frac{m_p\,cos^2\left(\theta\right)}{m_c\,+\,m_p}\right]}
\end{align}
>>>>>>>>>>>>>>>>>>>>>>>>$(*)$

\begin{align}
  \ddot{p}_c&=\frac{F\,+\,m_p\,l\,\left[\dot{\theta}^2\,\sin\left(\theta\right)\,-\,\ddot{\theta}\,\cos(\theta)\right]}{m_c\,+\,m_p}
\end{align}

where $g = 9.8\,\text{m/sec}$ is the gravitational acceleration, $m_c = 1\,\text{kg}$ is the cart mass, $m_p = 0.1\,\text{kg}$ is the pole mass, $l = 0.5\,\text{m}$ is the half-pole length, $\mu_p = 0.000002$ is the pole on cart friction coefficient.

```python
import numpy as np
from scipy.integrate import solve_ivp, odeint
from numpy import linalg as LA
import matplotlib.pyplot as plt
from scipy.io import loadmat
import jax.numpy as jnp
from jax import jacfwd
from jax import config
from scipy.optimize import fsolve
from IPython.display import HTML, clear_output
import matplotlib.animation as animation
from matplotlib import rc
import matplotlib
from tqdm import tqdm

matplotlib.rcParams['animation.embed_limit'] = 2**128
rc('animation', html='jshtml')
config.update("jax_enable_x64", True)
```

```python
class InvertedPendulum:
  def __init__(self, m_c, m_p, g, l, u_p):
    self.m_c = m_c
    self.m_p = m_p
    self.g = g
    self.l = l
    self.u_p = u_p

  ############################################################################
  ## Computation of the non-linear system
  ############################################################################

  def compute(self, t, x: np.array, u: float) -> tuple:
    dx = [None for _ in range(4)]
    dx[0] = x[1]
    dx[1] = (self.g*jnp.sin(x[0]) + jnp.cos(x[0])*((-u -self.m_p*self.l*jnp.sin(x[0])*x[1]**2)/(self.m_c + self.m_p)) - (self.u_p*x[1])/(self.m_p*self.l))/(self.l*(4/3 - (self.m_p*jnp.cos(x[0])**2)/(self.m_c + self.m_p)))
    dx[2] = x[3]
    dx[3] = (u + self.m_p*self.l*(jnp.sin(x[0])*x[1]**2 - dx[1]*jnp.cos(x[0])))/(self.m_c + self.m_p)

    dx = jnp.asarray(dx)

    return dx



  def render_cart_pole(self, time: float, state_evolution: np.array, sample_time: float):
    fig, ax = plt.subplots()

    cart_position = state_evolution[2, :]
    pole_angle = state_evolution[0, :]

    cart_width = 0.3
    cart_height = 0.2

    cart = plt.Rectangle((cart_position[0] - cart_width/2, 0), cart_width, cart_height, color='blue')
    pole_length = self.l*2
    pole = plt.Line2D([cart_position[0], cart_position[0] + pole_length*np.sin(pole_angle[0])],
                     [cart_height, cart_height + pole_length*np.cos(pole_angle[0])], color='black', linewidth=2)
    ax.set(xlim=[-2, 2], ylim=[-2, 2], xlabel='Position X [m]', ylabel='Position Y [m]')

    ax.grid()

    ax.add_patch(cart)
    ax.add_line(pole)

    def update(frame):
      cart.set_x(cart_position[frame] - cart_width/2)
      pole.set_data([cart_position[frame], cart_position[frame] + pole_length*np.sin(pole_angle[frame])],
                     [cart_height, cart_height + pole_length*np.cos(pole_angle[frame])])
      return (cart, pole)


    ani = animation.FuncAnimation(fig=fig, func=update, frames=state_evolution.shape[1], interval=sample_time*1000)
    plt.close()

    return ani
```

```python
def plot_solutions(t, solutions):
  # Create a figure with 2x2 subplots
  fig, axes = plt.subplots(1, 2, figsize=(10, 3))
  # Subplot 1 (left)
  axes[0].plot(t, solutions[0, :], label=r'theta')
  axes[0].set_title(r'$\theta(t)$')
  axes[0].legend()
  # Subplot 2 (right)
  axes[1].plot(t, solutions[0, :], label='theta')
  axes[1].plot(t, solutions[1, :], label='d(theta)')
  axes[1].plot(t, solutions[2, :], label='p')
  axes[1].plot(t, solutions[3, :], label='d(p)')
  axes[1].set_title(r'$\theta(t)$, $\dot{\theta}(t)$, $p(t)$, $\dot{p}(t)$')
  axes[1].legend()
  # Adjust layout
  plt.tight_layout()
  # Show the plot
  plt.show()
```

```python
inv_pend = InvertedPendulum(m_c=1, m_p=0.1, g=9.8, l=0.5, u_p=0.000002)
```

2. Provide a proper discretization of the state space $\mathcal{X}$, and define the size of the action-value function $Q$




*Answer: (Arguing the answer)*

Since the task is pole balancing a natural discretization takes into account the fact that we have to produce a denser set of state space values around the equilibrium point. A possible discretization is:

1. $\theta \in \left[-\frac{\pi}{4}, \frac{\pi}{4}\right]$ with step $\frac{\pi}{90}$ and $\theta \in \left[-\pi, -\frac{\pi}{4}\right] \cup \left[\frac{\pi}{4}, \pi \right]$ with step $\frac{\pi}{18}$

2. $\dot{\theta} \in \left[-10, 10 \right]$ with step $1$ $\forall k \in \mathbb{Z}$

3. $p \in \left[-10, 10 \right]$ with step $\frac{2}{5}$ $\forall k \in \mathbb{Z}$

4. $\dot{p} \in \left[-10, 10 \right]$ with step $1$ $\forall k \in \mathbb{Z}$

5. $F \in \left[-10, 10 \right]$ with step $1$ $\forall k \in \mathbb{Z}$

**Table size**

The $Q$ function table size is:

1. $\theta \in \left[-\frac{\pi}{4}, \frac{\pi}{4}\right]$ I have $46$ elements; for  $\theta \in \left[-\pi, -\frac{\pi}{4}\right] \cup \left[\frac{\pi}{4}, \pi \right]$ I have $14 + 14 - 2 = 26$ elements.

2. $\dot{\theta} \in \left[-10, 10 \right]$ I have $21$ elements.
3. $p \in \left[-10, 10 \right]$ I have $51$ elements.
4. $\dot{p} \in \left[-10, 10 \right]$ I have $21$ elements.
5. $F \in \left[-10, 10 \right]$ I have $21$ elements.

So the table is $Q = \bigl(( 46 + 26 )*21*51*21\bigr) \times 21$



```python
dstate_space = [
        np.hstack((np.linspace(-np.pi, -np.pi/8, 5, endpoint=False), np.linspace(-np.pi/8, np.pi/8, 10), np.linspace(np.pi/8, np.pi, 5, endpoint=False))),
        np.linspace(-1, 1, 21),
        np.linspace(-1, 1, 21),
        np.linspace(-1, 1, 21)
]

inputs = np.asarray([-10, 10])
number_of_inputs = inputs.shape[0]

q_table = np.zeros(np.hstack(([dstate_space[i].shape[0] for i in range(len(dstate_space))], [number_of_inputs])))
```

```python
q_table.shape # la tabella l'ho fatta in questo modo così ho già tutte le possibili combinazioni degli stati
```

```python
episodes = np.prod(q_table.shape)*2 # can't afford it with slow integration
episodes = 1000

episodes
```

```python
def discretize_state_index(state):
  dstate = [np.argmin(np.abs(state[i] - dstate_space[i])) for i in range(len(dstate_space))]
  return tuple(dstate)

discretize_state_index(np.array([np.pi/13, 1, -1, 1]))
```

```python
dstate_space[0]*(180/np.pi), dstate_space[1]
```

```python
POSSIBLE_STATE_LOW = [7, 10, 5, 10]
POSSIBLE_STATE_HIGH = [12, 11, 15, 11]

def get_random_state_index():
  return tuple(np.random.randint(low=POSSIBLE_STATE_LOW, high=POSSIBLE_STATE_HIGH))

get_random_state_index()
```

```python
def epsilon_greedy_policy(epsilon, state_idx):
  greedy_vs_explore = np.random.uniform(0,1)

  if greedy_vs_explore > epsilon: # Exploiting
    u = inputs[np.argmax(q_table[state_idx])]
  else: # Exploring
    u = inputs[np.random.randint(inputs.shape[0])]
  return u

epsilon_greedy_policy(0.2, discretize_state_index(np.array([np.pi/13, 1, -1, 1])))
```

```python
def greedy_policy(state_idx):
  return inputs[np.argmax(q_table[state_idx])]

greedy_policy(discretize_state_index(np.array([np.pi/13, 1, -1, 1])))
```

```python
MAX_ITERS = 1000

def compute_reward(state, iter):
  reward = 1

  if iter > MAX_ITERS:
    reward = 0
  elif abs(state[2]) > 0.5:
    reward = 0
  elif abs(state[0]) > np.pi/8:
    reward = 0

  return reward
```

```python
import time

def q_learning(alpha, gamma):
  total_returns = 0
  for episode in tqdm(range(episodes)):
    x_init_idx = get_random_state_index()
    x_init = np.array([dstate_space[i][x_init_idx[i]] for i in range(len(dstate_space))])

    done = False
    iter = 0
    episode_return = 0
    while not done:

      u = epsilon_greedy_policy(0.2, x_init_idx)
      u_idx = np.argmin(np.abs(inputs - u))

      sol = solve_ivp(fun=lambda t, x: inv_pend.compute(t, x, u), t_span=[0, 0.02], y0=x_init, t_eval=[0, 0.02], method='RK45', rtol=1e-3, atol=1e-6)
      sol = sol.y[:, 1]

      x_new_idx = discretize_state_index(sol)
      x_new = np.array([dstate_space[i][x_new_idx[i]] for i in range(len(dstate_space))])

      reward = compute_reward(x_new, iter)

      episode_return += reward

      fq_estim = np.max(q_table[x_new_idx])

      done = reward != 1

      if not done:
        q_table[x_init_idx][u_idx] = q_table[x_init_idx][u_idx] + alpha*(reward + gamma*fq_estim - q_table[x_init_idx][u_idx])

      x_init = x_new

      iter += 1

    total_returns += episode_return

    if episode % 10 == 0:
      print(f'mean returns {total_returns/10}')
      total_returns = 0
```

```python
q_learning(alpha=0.3, gamma=0.99)
```

```python
def simulate(x_init, simulation_time, sample_time) -> tuple:
  t_eval = np.linspace(0, simulation_time, int(simulation_time/sample_time) + 1)
  result = np.empty((x_init.shape[0], t_eval.shape[0]))
  result[:, 0] = x_init

  u_seq = np.empty(t_eval.shape[0] - 1)

  # simulation
  for i in range(0, t_eval.shape[0] - 1):

    x_init_idx = discretize_state_index(x_init)

    u = greedy_policy(x_init_idx)
    u_seq[i] = u

    sol = solve_ivp(fun=lambda t, x: inv_pend.compute(t, x, u), t_span=[0, 0.02], y0=x_init, t_eval=[0, 0.02], method='RK45', rtol=1e-3, atol=1e-6)
    x_init = sol.y[:, 1]

    result[:, i+1] = x_init

  return (t_eval, result, u_seq)
```

```python
t_s1, states_s1, u_seq = simulate(x_init=np.array([np.pi/13, 0, 0.1, 0]), simulation_time = 5, sample_time=0.02)
```

```python
def sarsa_learning(alpha, gamma):
  for episode in tqdm(range(episodes)):
    x_init_idx = get_random_state_index()
    x_init = np.array([dstate_space[i][x_init_idx[i]] for i in range(len(dstate_space))])

    done = False
    iter = 0

    u = epsilon_greedy_policy(0.2, x_init_idx)
    while not done:

      u_idx = np.argmin(np.abs(inputs - u))

      sol = odeint(func=lambda x, t: inv_pend.compute(t, x, u), t=[0, 0.02], y0=x_init)
      sol = sol[1]

      x_new_idx = discretize_state_index(sol)
      x_new = np.array([dstate_space[i][x_new_idx[i]] for i in range(len(dstate_space))])

      u_new = epsilon_greedy_policy(0.2, x_init_idx)
      u_new_idx = np.argmin(np.abs(inputs - u_new))

      reward = compute_reward(x_new, iter)

      done = reward != 1

      if not done:
        q_table[x_init_idx][u_idx] = q_table[x_init_idx][u_idx] + alpha*(reward + gamma*q_table[x_new_idx][u_new_idx] - q_table[x_init_idx][u_idx])

      x_init = x_new
      u = u_new

      iter += 1
```
